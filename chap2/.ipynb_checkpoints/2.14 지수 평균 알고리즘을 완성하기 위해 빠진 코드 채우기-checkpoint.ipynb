```python
import tensorflow as tf
import numpy as np
```

```python
raw_data = np.random.normal(10, 1, 100) # 평균값 10과 표준편차 1을 가지는 100개의 숫자로 구성된 벡터를 생성합니다

alpha = tf.constant(0.05) # alpha를 상수로 정의합니다
curr_value = tf.placeholder(tf.float32) # 플레이스홀더는 변수와 유사하지만 값이 세션으로부터 채워집니다
prev_avg = tf.Variable(0.) # 이전의 평균을 0으로 초기화합니다
update_avg = alpha * curr_value + (1 - alpha) * prev_avg

init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    for i in range(len(raw_data)): # 평균을 업데이트하기 위해 데이터 사이를 하나씩 돕니다
        curr_avg = sess.run(update_avg, feed_dict={curr_value: raw_data[i]})
        sess.run(tf.assign(prev_avg, curr_avg))
        print(raw_data[i], curr_avg)
```
