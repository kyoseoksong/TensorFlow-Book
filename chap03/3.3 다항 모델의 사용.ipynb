```python
import tensorflow as tf # 관련 라이브러리를 불러오고 하이퍼파라미터를 초기화합니다
import numpy as np
import matplotlib.pyplot as plt
```

```python
learning_rate = 0.01
training_epochs = 40

trX = np.linspace(-1, 1, 101) # 가짜 입력 데이터를 생성합니다

num_coeffs = 6 # 출력 데이터를 생성합니다
trY_coeffs = [1, 2, 3, 4, 5, 6]
trY = 0

for i in range(num_coeffs):
    trY += trY_coeffs[i] * np.power(trX, i)

trY += np.random.randn(*trX.shape) * 1.5 # 노이즈를 더해줍니다
```

```python
plt.scatter(trX, trY) # 원 데이터의 산포도를 그립니다
plt.show()
```

```python
X = tf.placeholder(tf.float32) # 입력/출력 쌍의 값을 저장할 노드를 정의합니다
Y = tf.placeholder(tf.float32)

def model(X, w): # 다항 모델을 정의합니다
    terms = []
    for i in range(num_coeffs):
        term = tf.multiply(w[i], tf.pow(X, i))
        terms.append(term)
    return tf.add_n(terms)

w = tf.Variable([0.] * num_coeffs, name="parameters") # 파라미터 벡터를 모두 0으로 설정합니다
y_model = model(X, w)

cost = (tf.pow(Y-y_model, 2)) # 이전과 마찬가지로 비용 함수를 정의합니다
train_op = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)

sess = tf.Session() # 이전과 마찬가지로 세션을 설정하고 학습 알고리즘을 실행합니다
init = tf.global_variables_initializer()
sess.run(init)

for epoch in range(training_epochs):
    for (x, y) in zip(trX, trY):
        sess.run(train_op, feed_dict={X: x, Y: y})

w_val = sess.run(w)
print(w_val)

sess.close() # 종료 시 세션을 닫아줍니다

plt.scatter(trX, trY) # 결과를 시각화합니다
trY2 = 0

for i in range(num_coeffs):
    trY2 += w_val[i] * np.power(trX, i)
    
plt.plot(trX, trY2, 'r')
plt.show()
```
